```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

from selenium import webdriver
import chromedriver_autoinstaller
```

```python
chromedriver_autoinstaller.install()
browser = webdriver.Chrome()
```

```python
browser.get('https://www.ruparupa.com/c/furniture.html')
```

```python
soup = BeautifulSoup(browser.page_source, "lxml")
```

```python
name_product = soup.find_all('span', class_='product__name')
```

```python
nama_produk = []
for each_name_product in name_product:
    nama_produk.append(each_name_product.text)
```

```python
nama_produk
```

```python
price_discount = soup.find_all('div', class_='price__discount')
```

```python
diskon_produk = []
for each_price_discount in price_discount:
    diskon_produk.append(each_price_discount.text)
```

```python
diskon_produk
```

```python
price_product = soup.find_all('div', class_='col-xs-12 price__real')
```

```python
harga_produk = []
for each_price_product in price_product:
    harga_produk.append(each_price_product.text)
```

```python
harga_produk
```

```python
data = []

for index in range (1,30):
    browser.get('https://www.ruparupa.com/c/furniture.html {index}')
    time.sleep(10)
    soup = BeautifulSoup(browser.page_source, "lxml")
    rows = soup.find_all('div', class_='row product-card')
    
    for row in rows:
        judul_produk = row.find(class_='product__name').text
        diskon_element = row.find(class_='price__discount')
        diskon = diskon_element.text if diskon_element is not None else ""
        sebelum_diskon_element = row.find(class_='price__initial')    
        sebelum_diskon = sebelum_diskon_element.text if sebelum_diskon_element is not None else ""
        setelah_diskon_element = row.find(class_='col-xs-12 price__real')
        setelah_diskon = setelah_diskon_element.text if setelah_diskon_element is not None else ""
        dt = {
            'judul_produk': judul_produk,
            'diskon': diskon,
            'sebelum_diskon': sebelum_diskon,
            'setelah_diskon': setelah_diskon,
            }
        
        data.append(dt)
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
nama_file = 'data_harga_furniture.csv'
df.to_csv(nama_file, index=False)
```
